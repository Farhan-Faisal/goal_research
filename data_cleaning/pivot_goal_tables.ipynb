```python
import pandas as pd
import os
```

```python
goal_path = "../data/proc/goals/daily_goal_embeddings_long.csv"
```

```python
cols = ["ParticipantIdentifier", "ResultIdentifier", "Answers", "trial_date", "label", "embedding"]
goal_df = pd.read_csv(goal_path)[cols].drop_duplicates(subset=['ParticipantIdentifier', 'ResultIdentifier', 'trial_date'])
goal_df.head()
```

```python
goal_wide = goal_df[["ParticipantIdentifier", "ResultIdentifier", "Answers", "trial_date"]].pivot(
    index=['ParticipantIdentifier', 'trial_date'],
    columns=['ResultIdentifier'], 
    values='Answers'
).reset_index()

label_wide = goal_df[["ParticipantIdentifier", "ResultIdentifier", "trial_date", "label"]].pivot(
    index=['ParticipantIdentifier', 'trial_date'],
    columns=['ResultIdentifier'], 
    values='label'
).rename(columns={
    "DAILY_goal1_set": "DAILY_goal1_label",
    "DAILY_goal2_set": "DAILY_goal2_label"
}).reset_index()

embedding_wide = goal_df[["ParticipantIdentifier", "ResultIdentifier", "trial_date","embedding"]].pivot(
    index=['ParticipantIdentifier', 'trial_date'],
    columns=['ResultIdentifier'], 
    values='embedding'
).rename(columns={
    "DAILY_goal1_set": "DAILY_goal1_embedding",
    "DAILY_goal2_set": "DAILY_goal2_embedding"
}).reset_index()
```

```python
goal_wide = goal_wide.merge(label_wide, on=["ParticipantIdentifier", "trial_date"])
goal_wide = goal_wide.merge(embedding_wide, on=["ParticipantIdentifier", "trial_date"])
```
